Analyzing Customer Orders Using Python




```python
#Data
#Customer list
customers = ["Alice", "Bob", "Charlie", "David", "Eva"]

#Orders stored as tuples inside a list
orders = [
    ("Alice", "Laptop", 800, "Electronics"),
    ("Alice", "T-Shirt", 20, "Clothing"),
    ("Bob", "Headphones", 50, "Electronics"),
    ("Bob", "Blender", 60, "Home"),
    ("Charlie", "Shoes", 120, "Clothing"),
    ("David", "Smartphone", 900, "Electronics"),
    ("David", "Watch", 200, "Electronics"),
    ("Eva", "Book", 15, "Home"),
    ("Eva", "Lamp", 35, "Home")
]

#list of products
customer_orders = {}

for customer, product, price, category in orders:
    if customer not in customer_orders:
        customer_orders[customer] = []
    customer_orders[customer].append(product)

print(customer_orders)
```

Product Classification

```python
#Category mapping
product_category = {}

for customer, product, price, category in orders:
    product_category[product] = category
```

```python
#Unique categories using a set
categories = set()

for _, _, _, category in orders:
    categories.add(category)

print("Product Categories:", categories)
```

Customer Spending Analysis

```python
#Calculate total spending per customer
customer_spending = {}

for customer, product, price, category in orders:
    if customer not in customer_spending:
        customer_spending[customer] = 0
    customer_spending[customer] += price
    print(price)
```

```python
#Classify customers
customer_classification = {}

for customer, total in customer_spending.items():
    if total > 100:
        customer_classification[customer] = "High-value buyer"
    elif 50 <= total <= 100:
        customer_classification[customer] = "Moderate buyer"
    else:
        customer_classification[customer] = "Low-value buyer"
print(customer_classification)
```

Business Insights

```python
#Total revenue per category
category_revenue = {}

for customer, product, price, category in orders:
    if category not in category_revenue:
        category_revenue[category] = 0
    category_revenue[category] += price

print(category_revenue)
```

```python
#Unique products
unique_products = set()

for _, product, _, _ in orders:
    unique_products.add(product)

print(unique_products)
```

```python
#Customers who bought Electronics
electronics_customers = [
    customer for customer, product, price, category in orders
    if category == "Electronics"
]

print(set(electronics_customers))
```

```python
#Top 3 highest spending customers
top_customers = sorted(customer_spending.items(), key=lambda x: x[1], reverse=True)[:3]

print(top_customers)
```

Set Operations & Insights

```python
#Customers who bought multiple categories
customer_categories = {}

for customer, product, price, category in orders:
    if customer not in customer_categories:
        customer_categories[customer] = set()
    customer_categories[customer].add(category)

multi_category_customers = [
    customer for customer, cats in customer_categories.items()
    if len(cats) > 1
]

print(multi_category_customers)
```

```python
#Customers who bought BOTH Electronics AND Clothing
both_categories = []

for customer, cats in customer_categories.items():
    if "Electronics" in cats and "Clothing" in cats:
        both_categories.append(customer)

print(both_categories)
```

FINAL REPORT OUTPUT

```python
#output
print("\n===== CUSTOMER SUMMARY =====")

for customer in customers:
    total = customer_spending.get(customer, 0)
    classification = customer_classification.get(customer, "No purchases")

    print(f"{customer}: ${total} -> {classification}")

print("\n===== CATEGORY REVENUE =====")
print(category_revenue)

print("\n===== UNIQUE PRODUCTS =====")
print(unique_products)

print("\n===== TOP 3 CUSTOMERS =====")
print(top_customers)

print("\n===== MULTI-CATEGORY CUSTOMERS =====")
print(multi_category_customers)

print("\n===== ELECTRONICS + CLOTHING BUYERS =====")
print(both_categories)
```
